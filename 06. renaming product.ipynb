```python
# import libraries
import numpy as np
import pandas as pd

# import CSV data
supsto = pd.read_csv("e:/SampleSuperstore.csv", encoding='windows-1252')

print(supsto.shape)
print(supsto.shape[0])

supsto.columns
```

```python
# convert date data types
supsto['ShipDate'] = pd.to_datetime(supsto['ShipDate'])
supsto['OrderDate'] = pd.to_datetime(supsto['OrderDate'])
```

```python
# find product with 'and' slang
supsto['informal_and'] = supsto['ProductName'].str.extract(r'(\s\W{1}[N|n])')
```

```python
supsto1 = supsto[['ProductName', 'informal_and']]
supsto1.head(5)
```

```python
# total product name with 'and' slang
supsto['informal_and'].count()
```

```python
# Showing product name with 'and' slang
selected_rows = supsto[supsto['informal_and'].notnull()]

selected_rows
```

```python
# Remove informal_and Column
supsto.drop('informal_and',axis=1, inplace=True)
```

```python
# replace 'And' slang to formal 'And'
supsto1 = supsto.replace({'ProductName': r'(\s\W{1}[N|n])'}, {'ProductName': ' And'}, regex=True)

supsto1.head()
```

```python
# Export to CSV
import os  

os.makedirs('folder/subfolder', exist_ok=True)  

supsto1.to_csv('e://superstore_modified.csv',index=False) 
```
